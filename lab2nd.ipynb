```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/drive/MyDrive/Iris.csv")
df
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.info()
```

```python
plt.figure(figsize=(6, 6))
df.boxplot(column=['SepalLengthCm'])
plt.title('Box Plot of Iris Dataset Features')
plt.ylabel('Measurement (cm)')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(6, 6))
df.boxplot(column=['SepalWidthCm'])
plt.title('Box Plot of Iris Dataset Features')
plt.ylabel('Measurement (cm)')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(6, 6))
df.boxplot(column=['PetalLengthCm'])
plt.title('Box Plot of Iris Dataset Features')
plt.ylabel('Measurement (cm)')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(6, 6))
df.boxplot(column=['PetalWidthCm'])
plt.title('Box Plot of Iris Dataset Features')
plt.ylabel('Measurement (cm)')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df.boxplot(column=['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm'])
plt.title('Box Plot of Iris Dataset Features')
plt.ylabel('Measurement (cm)')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df['PetalLengthCm'], label='Petal Length', fill=True)
plt.title('Univariate Density Plots of Iris Dataset Features')
plt.xlabel('Measurement (cm)')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df['SepalLengthCm'], label='Sepal Length', fill=True)
plt.title('Univariate Density Plots of Iris Dataset Features')
plt.xlabel('Measurement (cm)')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df['SepalWidthCm'], label='Sepal Width', fill=True)
plt.title('Univariate Density Plots of Iris Dataset Features')
plt.xlabel('Measurement (cm)')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df['PetalWidthCm'], label='Petal Width', fill=True)
plt.title('Univariate Density Plots of Iris Dataset Features')
plt.xlabel('Measurement (cm)')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df['SepalLengthCm'], label='Sepal Length', fill=True)
sns.kdeplot(data=df['SepalWidthCm'], label='Sepal Width', fill=True)
sns.kdeplot(data=df['PetalLengthCm'], label='Petal Length', fill=True)
sns.kdeplot(data=df['PetalWidthCm'], label='Petal Width', fill=True)
plt.title('Univariate Density Plots of Iris Dataset Features')
plt.xlabel('Measurement (cm)')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df['SepalLengthCm'], kde=True, color='blue', label='Sepal Length')
sns.histplot(data=df['SepalWidthCm'], kde=True, color='orange', label='Sepal Width')
sns.histplot(data=df['PetalLengthCm'], kde=True, color='green', label='Petal Length')
sns.histplot(data=df['PetalWidthCm'], kde=True, color='red', label='Petal Width')
plt.title('Univariate Histograms of Iris Dataset Features')
plt.xlabel('Measurement (cm)')
plt.ylabel('Frequency / Density')
plt.legend()
plt.show()
```
