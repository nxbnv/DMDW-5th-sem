```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/drive/MyDrive/student-mat.csv")
df
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
df.drop_duplicates(inplace=True)
```

```python
selected_attributes = ["famrel", "freetime", "goout", "Dalc","Walc","health", "absences", "G1", "G2", "G3"]
```

```python
correlation_matrix = df[selected_attributes].corr()
```

```python
plt.figure(figsize=(10, 8))


sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)


plt.title("Correlation Matrix Plot for Selected Attributes")

plt.show()
```

```python
sns.set(style="ticks")
sns.pairplot(df[selected_attributes], diag_kind="kde")


plt.suptitle("Scatter Matrix for Selected Attributes", y=1.02)


plt.show()
```
