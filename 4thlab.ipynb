```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr

# Load the dataset
df = pd.read_csv('/content/drive/MyDrive/student-mat.csv')

# Select the relevant columns for analysis
dfn = df[['traveltime', 'studytime']]

# Extract data from the DataFrame
x = dfn['traveltime']
y = dfn['studytime']


# Calculate correlation matrix
correlation_matrix = dfn.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix,cmap='plasma')
plt.title('Correlation Heatmap')
plt.show()

```

```python
dfn = df[['traveltime', 'studytime', 'age', 'failures', 'absences']]

# Calculate correlation matrix
correlation_matrix = dfn.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='PuRd')
plt.title('Correlation Heatmap')
plt.show()

# Print the correlation matrix
print("Correlation Matrix:")
print(correlation_matrix)



```

```python
binary_attributes = ['schoolsup', 'famsup']  # Replace with your binary attribute names

# Convert categorical attributes to numerical using one-hot encoding
dfn = pd.get_dummies(df[binary_attributes], columns=binary_attributes)

# Calculate correlation matrix
correlation_matrix = dfn.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

# Print the correlation matrix
print("Correlation Matrix:")
print(correlation_matrix)


```
